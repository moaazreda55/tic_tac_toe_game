```python
cur_player = 'X'
winner = None
gamerunnin = True
board=[' ',' ',' ',
             ' ',' ',' ',
              ' ',' ',' ']
def printBoard(board):
    print(board[0] + '|' + board[1] + '|' + board[2])
    print('-+-+-')
    print(board[3] + '|' + board[4] + '|' + board[5])
    print('-+-+-')
    print(board[6] + '|' + board[7] + '|' + board[8])
def DrawBoard(board):    
    print(" %c | %c | %c " % (board[0],board[1],board[2]))    
    print("___|___|___")    
    print(" %c | %c | %c " % (board[3],board[4],board[5]))    
    print("___|___|___")    
    print(" %c | %c | %c " % (board[6],board[7],board[8]))    
    print("   |   |   ")  
def playerinput(board):
    def eror():
        while True :
            global inp
            try :
                inp=int(input('select the position number'))
                break
            except ValueError :
                print('please enter a number')
        return True
    eror()
    while True :
        if inp in range(9) and board[inp]==' ' :
            board[inp] = cur_player
            break 
        else :
            print('this position is not available')
            eror()
def CheckWinner(board):    
    global winner    
    #Horizontal winning condition    
    if(board[0] == board[1] and board[1] == board[2] and board[0] != ' '):    
        winner = board[0]  
        return True 
    elif(board[3] == board[4] and board[4] == board[5] and board[3] != ' '):    
        winner = board[3]   
        return True 
    elif(board[6] == board[7] and board[7] == board[8] and board[6] != ' '):    
        winner = board[6]    
        return True 
    #Vertical Winning Condition    
    elif(board[0] == board[3] and board[3] == board[6] and board[0] != ' '):    
        winner = board[0]    
        return True 
    elif(board[1] == board[4] and board[4] == board[7] and board[1] != ' '):    
        winner = board[1]    
        return True 
    elif(board[2] == board[5] and board[5] == board[8] and board[2] != ' '):    
        winner=board[2]    
        return True 
    #Diagonal Winning Condition    
    elif(board[0] == board[4] and board[4] == board[8] and board[4] != ' '):    
        winner = board[4]    
        return True 
    elif(board[2] == board[4] and board[4] == board[6] and board[4] != ' '):    
        winner=board[4]    
        return True 
def CheckWin(board):
    global gamerunnin
    if CheckWinner(board) :
        DrawBoard(board)
        print(f'the winner is {winner}')
        gamerunnin = False
def checkdraw(board):
    global gamerunnin
    if ' ' not in board :
        DrawBoard(board)
        print('it is draw')
        gamerunnin = False  
def switch_player():
    global cur_player
    if cur_player == 'X' :
        cur_player = 'O'
    else :
        cur_player = 'X'
while gamerunnin :
    DrawBoard(board)
    playerinput(board)
    CheckWinner(board)
    CheckWin(board)
    checkdraw(board)
    switch_player()
        
```
